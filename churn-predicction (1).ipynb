```python
import numpy as np
import pandas as pd
import sklearn 
from sklearn import linear_model
```

```python
data = pd.read_csv("/kaggle/input/credit-card-customer-churn-prediction/Churn_Modelling.csv")
```

```python
data.head()
```

```python
data.drop(columns=["CustomerId", "Surname", "CustomerId"], inplace =True)
```

```python
data.head()
```

```python
data["Geography"].value_counts()
```

```python
data["Gender"].value_counts()
```

```python
data = pd.get_dummies(data,columns=['Geography','Gender'],drop_first=True)
```

```python
data.head()
```

```python
x = data.drop(columns = ["Exited"])
y= data["Exited"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

x_train_trf = scaler.fit_transform(x_train)
x_test_trf = scaler.transform(x_test)
```

```python
import tensorflow as tf 
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(12, activation ="relu", input_dim = 12))
model.add(Dense(5, activation= "relu"))
model.add(Dense(5, activation= "relu"))
model.add(Dense(1, activation = "sigmoid"))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, batch_size=50, epochs=100, verbose=1, validation_split=0.2)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred = y_pred.argmax(axis = -1)
```

```python
from sklearn.metrics import accuracy_score
```

```python
ACC = accuracy_score(y_test, y_pred)
```
